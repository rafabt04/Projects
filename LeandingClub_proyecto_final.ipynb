## Caso de negocio

Este caso analiza una plataforma de préstamos peer-to-peer (P2P) que conecta a prestatarios y prestamistas, eliminando intermediarios como bancos tradicionales, lo que puede llegar a ser muy llamativo para las personas al querer generar una ganancia por intereses, al igual que personas que requieren de liquidez inmediata sin tanta burocracia.

En este caso, se evaluan y aplican diversos temas incluyendo: conocimiento de riesgo crediticio, la evaluación de datos financieros y el uso de modelos predictivos de machine learning para tomar decisiones de inversión y poder validar la credibilidad de pago de los usuarios.

En este caso se usarán las siguientes variables:

 ## ***Variables Relacionadas con el Préstamo***

**loan_status**

Propósito: Indica el estado actual del préstamo (por ejemplo, Fully Paid, Current, Charged Off, etc.).
Relevancia: Variable objetivo clave para tareas de clasificación y predicción del desempeño del préstamo.


**loan_amnt**

Propósito: El monto total de dinero solicitado por el prestatario.
Relevancia: Impacta la exposición al riesgo del prestamista.


**installment**

Propósito: Monto del pago mensual del préstamo.
Relevancia: Útil para entender la carga financiera del prestatario y su capacidad de pago.


**int_rate**

Propósito: Tasa de interés asignada al préstamo.
Relevancia: Refleja el nivel de riesgo percibido (tasas más altas = mayor riesgo del prestatario).


**term**

Propósito: Periodo de reembolso del préstamo (36 meses o 60 meses).
Relevancia: Los términos más largos pueden aumentar el riesgo de incumplimiento debido a la exposición prolongada.


**funded_amnt**

Propósito: Monto que los inversionistas financiaron para el préstamo.
Relevancia: Muestra la confianza de los inversionistas en la solicitud del préstamo.



 ## ***Datos Demográficos y Solvencia del Prestatario***


**application_type**

Propósito: Indica si la solicitud es individual o conjunta.
Relevancia: Las solicitudes conjuntas pueden involucrar múltiples ingresos, afectando la capacidad de reembolso.


**home_ownership**

Propósito: Estado de la vivienda del prestatario (Alquiler, Propia, Hipotecada, etc.).
Relevancia: Una medida de estabilidad financiera; los propietarios suelen considerarse menos riesgosos.


**annual_inc**

Propósito: Ingreso anual del prestatario.
Relevancia: Evalúa la capacidad de reembolso y estabilidad financiera.


**verification_status**

Propósito: Indica si se verificaron los ingresos y el empleo del prestatario (Verified, Source Verified, o Not Verified).
Relevancia: Agrega confianza a la información financiera proporcionada.


**fico_range_high y fico_range_low**

Propósito: Límites superior e inferior del rango de puntajes FICO del prestatario.
Relevancia: Clave para evaluar la solvencia crediticia del prestatario.


 ## ***Métricas de Comportamiento Crediticio***


**revol_util**

Propósito: Porcentaje de utilización de crédito rotativo (proporción del límite de crédito utilizado).
Relevancia: Una alta utilización puede indicar estrés financiero.


**collections_12_mths_ex_med**

Propósito: Número de cobros en los últimos 12 meses, excluyendo deudas médicas.
Relevancia: Indica problemas financieros recientes y un mayor riesgo de incumplimiento.


**dti (Debt-to-Income Ratio)**

Propósito: Relación de las obligaciones mensuales de deuda con los ingresos mensuales.
Relevancia: Relaciones altas sugieren una capacidad limitada para asumir más deuda.


**open_acc**

Propósito: Número de cuentas de crédito activas actualmente.
Relevancia: Indica actividad crediticia y posible sobreendeudamiento.


**total_acc**

Propósito: Número total de cuentas de crédito que el prestatario ha tenido.
Relevancia: Proporciona contexto sobre el historial crediticio del prestatario.


**mo_sin_old_rev_tl_op**

Propósito: Meses desde que se abrió la cuenta rotativa más antigua.
Relevancia: Refleja la antigüedad del historial crediticio; los historiales más largos suelen indicar menor riesgo.


**total_bal_ex_mort**

Propósito: Saldo total de todas las cuentas de crédito, excluyendo hipotecas.
Relevancia: Indica la carga de deuda total y la exposición.

## Balancear etiquetas usando función resample

A continuación se describe el proceso de balanceo de etiquetas, el uso de las mismas y como resulta en la distribución después de balancearlas.

```python
from sklearn.utils import resample
#Crear dos diferentes dataframe de una clase mayoritaria y minoritaria
df_majority = df_cleaned[(df_cleaned['loan_status']==0)] 
df_minority = df_cleaned[(df_cleaned['loan_status']==1)] 

# muestreo ascendente de la clase minoritaria
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # muesta con reemplazo 
                                 n_samples= 1438202, # para que coincida con la clase mayoritaria
                                 random_state=0)   # resultados reproducible

# Combinar la clase mayoritaria con la muestra ascendente de la clase minoritaria 
df_upsampled = pd.concat([df_minority_upsampled, df_majority])
```

```python
# Calculamos los valores de cada categoría
loan_status_counts_upsampled = df_upsampled['loan_status'].value_counts()

# Creamos el gráfico de barras
plt.bar(loan_status_counts_upsampled.index, loan_status_counts_upsampled.values)

# Personalizamos el gráfico
plt.xticks(ticks=[0, 1], labels=['Fully Paid', 'Default'])  # Etiquetas de las barras
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.title('Distribución de Loan Status')

# Mostramos el gráfico
plt.show()

df_upsampled['loan_status'].value_counts()
```

```python
def remove_outliers_iqr(df_cleaned, columns):
    for col in columns:
        q1 = df_cleaned[col].quantile(0.25)
        q3 = df_cleaned[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
    return df_cleaned

# Señale las columnas para remover los outliers
columns_to_check = [
    'annual_inc', 'loan_amnt', 'installment', 'int_rate', 'fico_range_high', 
    'fico_range_low', 'revol_util', 'collections_12_mths_ex_med', 
    'funded_amnt', 'dti', 'open_acc', 'total_acc', 
    'mo_sin_old_rev_tl_op', 'total_bal_ex_mort'
]

# Eliminar columnas duplicadas en caso de que existan
df_upsampled = df_upsampled.loc[:, ~df_upsampled.columns.duplicated()]

# Revisar y manejar valores nulos antes de eliminar outliers
df_upsampled[columns_to_check] = df_upsampled[columns_to_check].dropna()

# Aplicar la función remove_outliers_iqr después de limpiar columnas duplicadas y manejar valores nulos
df_clean = remove_outliers_iqr(df_upsampled, columns_to_check)

# Mostrar el resultado
df_clean.head()
```

## Decision Tree

Al tener los resultados del accuracy score podemos ver que es mejor que el de regresión logística, ya que este nuevo modelo presenta un 71% exactitud.

```python
from sklearn.tree import DecisionTreeClassifier

# Configuración con un único criterio
dt_model = DecisionTreeClassifier(max_depth=20, min_samples_split=2, min_samples_leaf=5, criterion='gini', random_state=42)
dt_model.fit(X_train, y_train)
```

```python
#Hacemos predicciones
y_pred_dt = dt_model.predict(X_test)
```

```python
print('Precisión en el set de Entrenamiento: {:.2f}'
     .format(dt_model.score(X_train, y_train)))
print('Precisión en el set de Test: {:.2f}'
     .format(dt_model.score(X_test, y_test)))

print (classification_report(y_test, y_pred_dt))
```

## Métricas Decision Tree

Aquí podemos observar que a pesar de que la exactitud es buena, tenemos un modelo que puede tener errores de clasificación por los lo mostrado en la mariz de confusión. 

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
print('F-1 Score : ',(f1_score(y_test, y_pred_dt, average='micro')))
print('Precision Score : ',(precision_score(y_test, y_pred_dt, average='micro')))
print('Recall Score : ',(recall_score(y_test, y_pred_dt, average='micro')))
```

```python
# Matriz de confusión
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_dt)
cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='Reds')
plt.figure(figsize=(9,9))
```

## Modelos Elegidos

```python
from sklearn.metrics import log_loss
```

```python
algos = [LR_lasso,rf_model, dt_model, xgb_model]
ml_algo = ['Logistic Reg','Random Forest', "Decision Tree", 'XGBoost']

# Hacer un loop para todos los algoritmos usados en el problema de predicción
for i,j in zip(algos,ml_algo):
    i.fit(X_train,y_train)
    pred = i.predict(X_test)
    print(j,':\n')
    print('Accuracy Score:{:.2f}%'.format(accuracy_score(y_test,pred)*100))
    print('Log Loss:{:.2f}'.format(log_loss(y_test,pred)))
    print('='*40)
```

Para resumir los resultados y poder dar una recomencación del uso de los modelos, se debe a las caracteristicas del accuracy score, que nos dice la exactitud de de la predicción de los resultados por cada uno de los modelos, y el log loss, lo que nos indica la cantidad de información o bien resultados perdidos.

Teniendo en cuenta que queremos que el accuracy score sea lo mas cercano a 100% y que el log loss sea lo mas bajo posible, entonces pudieramos recomendar el modelo de random forest como el **primordial** por su 84.49% de accuracy y su log loss de 5.59.

También es importante poder dar a conocer que este modelo también es recomendable por temas de tiempo del procesamiento de la clasificación.
